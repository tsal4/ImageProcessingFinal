```python
import matplotlib.pyplot as plt
import numpy as np

from skimage import io, color
from skimage.morphology import disk, opening, closing, remove_small_objects
from skimage.measure import label, regionprops
from scipy.ndimage import binary_fill_holes
from matplotlib.patches import Rectangle

# -----------------------------
# LOAD IMAGE 008
# -----------------------------

img = io.imread("Wood/000.png")   # change filename if needed

# Remove alpha channel if present
if img.ndim == 3 and img.shape[2] == 4:
    img = img[:, :, :3]

# Convert RGB -> HSV
hsv = color.rgb2hsv(img)

hue = hsv[:, :, 0]
sat = hsv[:, :, 1]
val = hsv[:, :, 2]

# -----------------------------
# RED DEFECT DETECTION
# -----------------------------
# Red hues occur near 0 and 1 in HSV space

red_mask = (
    ((hue < 0.05) | (hue > 0.95)) &
    (sat > 0.18) &
    (val > 0.15)
)

# -----------------------------
# CLEANUP
# -----------------------------

cleaned = opening(red_mask, disk(2))
cleaned = closing(cleaned, disk(5))

cleaned = binary_fill_holes(cleaned)

cleaned = remove_small_objects(
    cleaned,
    min_size=150
)

# -----------------------------
# DISPLAY MASK RESULTS
# -----------------------------

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original Wood Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(red_mask, cmap='gray')
plt.title("Initial Red Detection")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cleaned, cmap='gray')
plt.title("Cleaned Defect Mask")
plt.axis("off")

plt.tight_layout()
plt.show()

# -----------------------------
# DRAW DEFECT BOXES
# -----------------------------

labeled_mask = label(cleaned)
regions = regionprops(labeled_mask)

fig, ax = plt.subplots(figsize=(8,8))

ax.imshow(img)
ax.set_title("Detected Red Paint Defects")
ax.axis("off")

defect_count = 0

for region in regions:

    if region.area >= 150:

        minr, minc, maxr, maxc = region.bbox

        rect = Rectangle(
            (minc, minr),
            maxc - minc,
            maxr - minr,
            linewidth=3,
            edgecolor='red',
            facecolor='none'
        )

        ax.add_patch(rect)

        ax.text(
            minc,
            minr - 10,
            "Red Paint Defect",
            color='red',
            fontsize=11,
            weight='bold'
        )

        defect_count += 1

plt.show()

print("Detected red paint defect regions:", defect_count)
```
